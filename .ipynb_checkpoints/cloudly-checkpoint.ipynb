Sağa çarpık dağılımlar, çok sayıda kişinin daha düşük maaş aldığı, ancak az sayıda kişinin aşırı yüksek maaş aldığı durumlarda ortaya çıkar.

2. Mod, Medyan ve Ortalama Arasındaki Farklar:
Bu üç değerin büyüklük farkları, verinin dağılımı hakkında bilgi verir:

Eğer mod, medyan ve ortalama birbirine yakınsa, veri normal dağılmış olabilir.
Eğer ortalama medyandan büyükse (ve sağa kaymışsa), veri sağa çarpık demektir (pozitif çarpıklık).
Eğer ortalama medyandan küçükse (ve sola kaymışsa), veri sola çarpık demektir (negatif çarpıklık).

```python
# salary_in_usd sütunu için mod, medyan ve ortalama değerlerini alalım
salary_mode = df['salary_in_usd'].mode()[0]
salary_median = df['salary_in_usd'].median()
salary_mean = df['salary_in_usd'].mean()

print(f"Mod: {salary_mode}, Medyan: {salary_median}, Ortalama: {salary_mean}")
```

```python
# Salary dağılımını ve istatistiksel değerleri gösterecek grafik
plt.figure(figsize=(10, 6))
sns.histplot(df['salary_in_usd'], kde=True, bins=30, color='blue', label='Salary Distribution')

# Ortalama, Medyan ve Modu gösterelim
plt.axvline(salary_mean, color='red', linestyle='--', label=f'Mean: {salary_mean:.2f}')
plt.axvline(salary_median, color='green', linestyle='-', label=f'Median: {salary_median:.2f}')
plt.axvline(salary_mode, color='purple', linestyle='-', label=f'Mod: {salary_mode:.2f}')

plt.title('Salary Distribution with Mean, Median, and Mode')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

3. Çarpıklık (Skewness) ve Dağılımın Yönü:
Çarpıklık, verinin asimetrisini ölçer. Verinin sağa çarpık mı, sola çarpık mı olduğunu anlamak için skewness (çarpıklık) değerini kullanırız:

Skewness > 0: Sağa çarpık (pozitif çarpıklık) — uzun kuyruk sağdadır.
Skewness < 0: Sola çarpık (negatif çarpıklık) — uzun kuyruk soldadır.
Skewness = 0: Simetrik dağılım (normal dağılım).

```python
# salary_in_usd sütunundaki çarpıklığı (skewness) hesaplayalım
salary_skewness = df['salary_in_usd'].skew()
print(f"Salary in USD Skewness: {salary_skewness:.2f}")
```

```python
# Salary in USD dağılımındaki çarpıklığı görselleştirelim
plt.figure(figsize=(10, 6))
sns.histplot(df['salary_in_usd'], kde=True, bins=30, color='blue')
plt.title(f'Salary in USD Distribution (Skewness: {salary_skewness:.2f})')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.show()
```

4. Verinin Çarpıklığını Yorumlama:
Pozitif çarpıklık: Maaş dağılımı sağa doğru uzuyor olabilir, yani birkaç kişi çok yüksek maaş alıyor.
Negatif çarpıklık: Maaş dağılımı sola doğru uzuyor olabilir, yani birkaç kişi çok düşük maaş alıyor.
Simetrik: Dağılım dengeli ise, maaşlar homojen bir şekilde dağılmış olabilir.

```python
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers
```

```python
# Maaş sütununda aykırı değerleri tespit et
salary_outliers = detect_outliers(df_cleaned, 'salary_in_usd')

print(f"Tespit edilen aykırı değer sayısı: {len(salary_outliers)}")
```

```python
# Aykırı değerleri düzeltme (winsorization yöntemi)
def winsorize(df, column, lower_percentile=0.05, upper_percentile=0.95):
    lower = df[column].quantile(lower_percentile)
    upper = df[column].quantile(upper_percentile)
    df[column] = df[column].clip(lower, upper)
    return df

df_winsorized = winsorize(df_cleaned.copy(), 'salary_in_usd')
```

```python
# Düzeltilmiş veri setini görselleştirme
plt.figure(figsize=(12, 6))
sns.boxplot(x='simplified_job_title', y='salary_in_usd', data=df_winsorized)
plt.title('Düzeltilmiş Maaş Dağılımı')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Bu analizler, sektördeki uzun vadeli trendleri anlamak, gelecekteki iş gücü ihtiyaçlarını tahmin etmek ve kariyer planlaması yapmak için çok değerli olabilir. Örneğin, bir şirket bu trendlere bakarak gelecekteki işe alım stratejilerini belirleyebilir veya bir profesyonel, kariyerini hangi yönde geliştirmesi gerektiğine karar verebilir.

```python
# 1. Null değerleri tespit et ve doldur
print("1. Null değerler:")
print(df.isnull().sum())

# Null değerleri doldur
numeric_columns = df.select_dtypes(include=[np.number]).columns
categorical_columns = df.select_dtypes(include=[object]).columns

imputer_numeric = SimpleImputer(strategy='median')
imputer_categorical = SimpleImputer(strategy='most_frequent')

df[numeric_columns] = imputer_numeric.fit_transform(df[numeric_columns])
df[categorical_columns] = imputer_categorical.fit_transform(df[categorical_columns])

print("\nNull değerler doldurulduktan sonra:")
print(df.isnull().sum())

# 2. Duplicate verileri tespit et ve temizle
duplicate_count = df.duplicated().sum()
print(f"\n2. Tespit edilen duplicate veri sayısı: {duplicate_count}")
df.drop_duplicates(inplace=True)
print(f"Duplicate veriler temizlendikten sonra kalan veri sayısı: {len(df)}")

# 3. Sayısal sütunlarda negatif değerleri tespit et ve düzelt
negative_counts = (df[numeric_columns] < 0).sum()
print("\n3. Negatif değerler:")
print(negative_counts)

for col in numeric_columns:
    df[col] = df[col].abs()  # Negatif değerleri pozitife çevir

# 4. Aykırı değerleri bul ve düzenle
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].clip(lower_bound, upper_bound)
    return df

print("\n4. Aykırı değerler düzenleniyor...")
for col in numeric_columns:
    df = remove_outliers(df, col)

# 5. Job title sütununu sadeleştir ve ilk 25 iş unvanlarını göster
def simplify_job_title(title):
    if 'data scientist' in title.lower():
        return 'Data Scientist'
    elif 'data analyst' in title.lower():
        return 'Data Analyst'
    elif 'machine learning' in title.lower():
        return 'Machine Learning Engineer'
    elif 'data engineer' in title.lower():
        return 'Data Engineer'
    elif 'ai' in title.lower() or 'artificial intelligence' in title.lower():
        return 'AI Specialist'
    elif 'business intelligence' in title.lower() or 'bi' in title.lower():
        return 'Business Intelligence'
    else:
        return 'Other'

df['simplified_job_title'] = df['job_title'].apply(simplify_job_title)

print("\n5. İlk 25 sadeleştirilmiş iş unvanı:")
print(df['simplified_job_title'].value_counts().head(25))

# Temizlenmiş veri setinin özeti
print("\nTemizlenmiş veri setinin özeti:")
print(df.info())

# İlk birkaç satırı göster
print("\nTemizlenmiş veri setinin ilk birkaç satırı:")
print(df.head())
```
